```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create data as dictionary
data = {
    "Order_ID": [101, 102, 103, 104, 105],
    "Date": ["2024-01-05", "2024-01-08", "2024-01-10", "2024-01-15", "2024-02-02"],
    "Product": ["Laptop", "Phone", "Shoes", "Watch", "Laptop"],
    "Category": ["Electronics", "Electronics", "Fashion", "Accessories", "Electronics"],
    "Quantity": [2, 3, None , 2, 1],
    "Price": [55000, 35000, None , 4000, 60000],
    "Region": ["North", "South", "East", "West", "North"],
    "Salesperson": ["Ayesha", "Bilal", "Ali", "Sana", "Ayesha"],
    "Payment_Mode": ["Credit Card", "Cash", "Credit Card", "UPI", "Credit Card"]
}

# Convert dictionary → DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
print(df.head(3))
```

```python
df.info() #shows columns, data types, and non-null counts.
```

```python
df.describe() #gives statistical summary (mean, min, max, etc.)
```

```python
df.isnull().sum() #helps detect missing data.
```

```python
df["Total_Sales"] = df["Quantity"] * df["Price"]
#creates a new column called Total_Sales in DataFrame
```

```python
df["Total_Sales"]
```

```python
df
```

```python
#Useful to compare performance of different product types.
category_sales = df.groupby("Category")["Total_Sales"].sum()  #category-wise sale
print(category_sales)
```

```python
Region_sales = df.groupby("Region")["Total_Sales"].sum()   #Region-wise sale
print(Region_sales)
```

```python
sales_man = df.groupby("Salesperson")["Total_Sales"].sum()
print(sales_man)
```

```python
category_sales.plot(kind="bar", color="yellow")  #.plot() function is pandas’ built-in shortcut for creating charts using matplotlib behind the scenes.
#It works directly on Series or DataFrames.
plt.title("Category-wise Total Sales")    #---charts kind = line , barh , hist , pie , box , area , scatter , kde
plt.xlabel("Category")
plt.ylabel("Sales (PKR)")
plt.show()

```

```python
#autopct='%1.1f%%' --This controls how the percentage values appear on the chart.
#figsize=(6,6)--sets the size of the figure (width, height) in inches.

colors = ['skyblue', 'lightgreen', 'lightcoral', 'gold', 'pink']

Region_sales.plot(
    kind="pie",
    autopct='%1.1f%%',
    figsize=(6,6),
    colors=colors   # color list
)
plt.title("Region-wise Sales Distribution")
plt.show()

```

```python
import seaborn as sns             # For advanced, stylish visualizations
#Seaborn is a Python library used for creating beautiful, informative, and easy-to-read visualizations with just a few lines of code.
```

```python
sns.barplot(x="Salesperson", y="Total_Sales", data=df, palette="viridis") #x=column ; y=numeric value ,data=df → tells Seaborn which DataFrame to use (df)
plt.title("Salesperson-wise Performance") #"viridis" , "cool" ,
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month_name()

month_sales = df.groupby("Month")["Total_Sales"].sum()
month_sales.plot(kind="bar", color="orange")
plt.title("Month-wise Sales Trend")
plt.ylabel("Total Sales")
plt.show()
```
