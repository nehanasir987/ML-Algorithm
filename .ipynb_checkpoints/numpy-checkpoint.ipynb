```python
import numpy as np                    #NumPy arrays (np.array) are faster, more memory-efficient, and more powerful than regular Python lists - matical structure use to crate a numbers
                                      #NumPy array is homogeneous, meaning all elements must be of the same data type
                                      #NumPy keeps exact numbers - take less memory
# deals with a number , matrices , scientific calculation
#

a = np.array([1, 2, 3],dtype=np.int8)               # list
b = np.array((49, 59, 65))               # tuple
c = np.array([[1,2,3],[4,5,6]])       # nested list = 2D
d = np.array(range(5))                # range object
```

```python
import numpy as np
arr1 = np.array([1, 2, 3, 4])
print(arr1.ndim)    # Output: 1    #ndim --> number of dimensions


arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2.ndim)  # Output: 2

arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(arr3.ndim)  # Output: 3

```

```python
zeros = np.zeros((2 , 5))
print(zeros , zeros.ndim , zeros.shape , zeros.size )
```

```python
print(zeros + 5)    ## Har element me 5 add
print(zeros * 10)  # Har element × 10
zeros[0, 2] = 99   # Pehli row, 3rd element = 99
print(zeros)

print(zeros.max())
print(zeros.min())

print(zeros[0])     # Pehli row
reshaped = zeros.reshape(5, 2)  # 2x5 → 5x2
print(reshaped)


ones = np.ones((2, 5))     # sab 1s
print(ones)
randoms = np.random.rand(2, 5)  # random numbers (0–1)
print(randoms)



```

```python
import numpy as np

print(np.random.rand(2,5))       # 0–1 ke beech float
print(np.random.randn(2,5))      # normal distribution
print(np.random.randint(10,50,(2,5)))  # 10 se 50 ke beech integers
print(np.random.random((2,5)))   # same as rand
```
