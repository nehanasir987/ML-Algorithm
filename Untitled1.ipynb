```python
'''Why We Use LabelEncoder

Machine Learning models (like regression, KNN, SVM, etc.)
cannot understand text data directly.
They only understand numbers.

So if you have a dataset like this:
Then ML model can’t process “setosa”, “versicolor”, “virginica”.

We use LabelEncoder to convert them into numbers: sitoca: 0 , vercical: 1  , viginica:2'''
```
