```python
import jwt
import datetime

# Secret key (like in Django settings)
SECRET_KEY = "mysecretkey987"

# Create a token (for login)
payload = {
    "user_id": 1,
    "username": "neha",
    "exp": datetime.datetime.utcnow() + datetime.timedelta(minutes=2)
}

token = jwt.encode(payload, SECRET_KEY, algorithm="HS256")
print("üîê Generated Token:\n", token)

# Decode token
decoded = jwt.decode(token, SECRET_KEY, algorithms=["HS256"])
print("\n‚úÖ Decoded Payload:\n", decoded)
```
